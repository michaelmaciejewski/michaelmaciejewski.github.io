```python
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import datetime
```

```python
df = pd.read_csv("the_office_series.csv")
```

```python
df.head()
```

```python
dfviewership = df.sort_values(by = ["Viewership","EpisodeTitle"], ascending = False).iloc[:15]
dfviewership.plot(kind="bar" ,x = 'EpisodeTitle', y = 'Viewership')
plt.ylabel("Amount of Views (in thousands)")
plt.xlabel("Episode Name")
plt.title("Top 15 Watched Episodes of The Office")
plt.savefig('my_plotnew.png', bbox_inches='tight')
```
